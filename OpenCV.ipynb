```python
import cv2
import tensorflow.keras
import numpy as np
import os
from gtts import gTTS
import playsound
import os
import time
import random

def speak(text):
    tts = gTTS(text=text, lang='en')
    filename='voice.mp3'
    tts.save(filename)
    playsound.playsound(filename)

def preprocessing(frame):
    size = (64, 64)
    frame_resized = cv2.resize(frame, size, interpolation=cv2.INTER_AREA)
    
    frame_normalized = (frame_resized.astype(np.float32) / 127.0) - 1
    
    frame_reshaped = frame_normalized.reshape((1, 64, 64, 3))
    
    return frame_reshaped

model_filename = 'conkeras_model (2).h5'
model = tensorflow.keras.models.load_model(model_filename)

capture = cv2.VideoCapture(0) # 비디오 캡쳐 객체를 생성

capture.set(cv2.CAP_PROP_FRAME_WIDTH, 320)
capture.set(cv2.CAP_PROP_FRAME_HEIGHT, 240)

#["coronacontent_ex", "docontent_ex", "fightingcontent_ex","preventioncontent_ex","normal"]
#   1                       2                3                   4                   5
cnt = 1 
while True:
    ret, frame = capture.read()
#     if ret == True: 
#         print("read success!")

    frame_fliped = cv2.flip(frame, 1)
    
    # 이미지 출력
    cv2.imshow("Signal2021", frame_fliped)
    
    if cv2.waitKey(200) > 0: 
        break
    
    preprocessed = preprocessing(frame_fliped)

    prediction = model.predict(preprocessed)
#     print(prediction) # [[0.00533728 0.99466264]]
#     print(prediction[0,0])
#     print(prediction[0,1])
#     print(prediction[0,2])
#     print(prediction[0,3])
    
    if prediction[0,0] < prediction[0,1] :
        print("코로나")
        time.sleep(random.uniform(0.1, 1))
#         speak("코로나")
#         os.remove(r"voice.mp3")
        cnt +=1
        #break
    elif prediction[0,1] < prediction[0,2]:
        print("합시다")
        time.sleep(random.uniform(0.1, 1))
#         speak("예방")
#         os.remove(r"voice.mp3")
        cnt +=1
        #break
    elif prediction[0,2] < prediction[0,3]:
        print("파이팅")
        time.sleep(random.uniform(0.1, 1))
#         speak("합시다")
#         os.remove(r"voice.mp3")
        cnt +=1
        #break
    elif prediction[0,3] <= prediction[0,3]:
        print("예방")
        time.sleep(random.uniform(0.1, 1))
#         speak("파이팅")
#         os.remove(r"voice.mp3")
        cnt +=1
        #break
    #elif prediction[0,4] <= prediction[0,4]:
     #   print("normal",end='\r')
      #  time.sleep(random.uniform(0.1, 1))
       # cnt +=1
        #break
    else:
        print("다른 동작을 실행해주세요")
        time.sleep(random.uniform(0.1, 1))
        if cnt % 50 == 0:
            cnt=1
            break

capture.release() # 오픈한 캡쳐 객체를 해제
cv2.destroyAllWindows() #윈도우 종료
```
